```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
token=Tokenizer()
token.fit_on_texts([corpus])
total_words=len(token.word_index) + 1
```

```python
# token.word_index
```

```python
input_sequence=[]
for line in corpus.split('\n'):
    token_list=token.texts_to_sequences([line])[0]

    for i in range(1,len(token_list)):
        n_gram=token_list[:i+1]
        input_sequence.append(n_gram)

```

```python
# input_sequence
```

```python
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences
max_len=max([len(x) for x in input_sequence])
input=np.array(pad_sequences(input_sequence,maxlen=max_len,padding='pre'))
```

```python
input
```

```python
X=input[:,:-1]
y=input[:,-1]
```

```python
y=np.array(tf.keras.utils.to_categorical(y,num_classes=total_words))
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Embedding,Dense
```

```python
model=Sequential()
```

```python
model.add(Embedding(total_words,100))
model.add(LSTM(150))
model.add(Dense(total_words,activation='softmax'))
          
```

```python
model.compile(optimizer='Adam',loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X,y,epochs=100,verbose=1)
```

```python
model.summary()
```

```python
input_text="india"
next_sequence=3
for i in range(next_sequence):
    token_list=token.texts_to_sequences([input_text])[0]
    token_list=pad_sequences([token_list],maxlen=max_len-1,padding='pre')
    predicted=np.argmax(model.predict(token_list))
    out_put=token.index_word[predicted]
    input_text+=" "+out_put
print(input_text)
```
